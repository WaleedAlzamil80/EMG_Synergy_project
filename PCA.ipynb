```python
scaler1=StandardScaler()
scaler1.fit(emg1)
scaled_data1=scaler1.transform(emg1)

#Initializes a StandardScaler object scaler, which is used for feature scaling. StandardScaler scales the input features to have zero mean and unit variance.
#Fits the scaler object to the data in df, using the fit method. This computes the mean and standard deviation of each feature in the input data.
#Applies the transform method of the scaler object to the df DataFrame. This scales the features of the input data using the mean and standard deviation computed in the previous step



pca1=PCA(n_components=12)
pca1.fit(scaled_data1)#pca.fit(scaled_data) fits the PCA object to the scaled data. This computes the principal components from the scaled data and stores them in the PCA object.

x_pca1=pca1.transform(scaled_data1)#pca.transform(scaled_data) applies the dimensionality reduction by transforming the scaled data into a lower-dimensional space using the principal components. The resulting transformed data is assigned to the variable x_pca, which is a numpy array with shape (n_samples, n_components).

```

```python
plt.figure(figsize = (8, 8))
plt.bar(range(1, len(pca1.explained_variance_ratio_) +  1), pca1.explained_variance_ratio_)
plt.xlabel("NO.Components")
plt.ylabel("Variance Captured by each component")
plt.title("Variance Explained")
plt.show()
```

```python
var_cum = np.cumsum(pca1.explained_variance_ratio_)

plt.figure(figsize = (8, 8))
plt.bar(range(1, len(var_cum) + 1), var_cum)
plt.xlabel("NO.Components")
plt.ylabel("Variance Captured by these number of components")
plt.title("Cumulative Variance Explained")
plt.show()
```
